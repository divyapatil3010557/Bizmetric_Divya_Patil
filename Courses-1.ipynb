```python
class DegreeFee:
    def __init__(self):
        self.core_fee=200000
        self.analytics_extra_percent=0.10
        self.hostel_fee_per_year=200000
        self.food_per_month=2000
        self.transportation_per_semester=13000
        self.l1=['HR', 'FINANCE', 'MARKETING', 'DS']
        self.subject=""
        self.analytics=""
        self.hostel=""
        self.food_months=0
        self.transport=""
        self.analytics_fee=0
        self.course_fee=0
        self.hostel_fee=0
        self.food_fee=0
        self.transport_fee=0
        self.total_cost=0
    def take_input(self):
        self.subject=input("Which course you want to pursue (HR, Finance, Marketing, DS): ").upper()
        if self.subject not in self.l1:
            print("Course is not available")
            exit()
        self.analytics=input("Do you want analytics? (Y/N): ").upper()
        self.hostel=input("Do you want hostel? (Y/N): ").upper()
        self.food_months=int(input("Enter number of food months: "))
        self.transport=input("Transportation (SEMESTER/ANNUAL): ").upper()
    def calculate_fees(self):
        self.course_fee=self.core_fee
        if self.subject!= "DS" and self.analytics =='Y':
            self.analytics_fee =self.core_fee*self.analytics_extra_percent
            self.course_fee += self.analytics_fee
        else:
            self.analytics_fee=0
        if self.hostel=='Y':
            self.hostel_fee=self.hostel_fee_per_year
        else:
            self.hostel_fee=0
        self.food_fee=self.food_per_month*self.food_months
        if self.transport=='SEMESTER':
            self.transport_fee=self.transportation_per_semester
        elif self.transport=='ANNUAL':
            self.transport_fee=self.transportation_per_semester*2
        else:
            print("Invalid transport option")
            exit()
        self.total_cost=(self.course_fee+self.hostel_fee+self.food_fee+self.transport_fee)
    def print_bill(self):
        print("Course Selected     :",self.subject) 
        print("Analytics Fee (10%) :",self.analytics_fee)
        print("Course Fee          :",self.course_fee) 
        print("Hostel Fee          :",self.hostel_fee) 
        print("Food Fee            :",self.food_fee) 
        print("Transportation Fee  :",self.transport_fee)
        print("-"*50) 
        print("Total Annual Cost   :",self.total_cost)
obj = DegreeFee()
obj.take_input()
obj.calculate_fees()
obj.print_bill()
```
