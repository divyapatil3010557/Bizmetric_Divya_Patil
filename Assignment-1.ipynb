#rating received as A, B, C, D
#If salary upto 5 lpa then increament based on ratings
#are  A = 16% , B=12%, C= 10%, D=6%
#If salary upto 10 lpa then increament based on ratings
#are  A = 14% , B=10%, C= 8%, D=6%
#If salary upto 15 lpa then increament based on ratings
#are  A = 8% , B=6%, C= 4%, D = no increment
#If salary upto 23 lpa then increament based on ratings
#are  A = 7% , B=5%, C= 4%, D=no

```python
salary = input("Enter your salary: ")
rating = input("Enter rating(A/B/C/D): ").upper()
if salary.replace('.', '', 1).isdigit():
    salary = float(salary)
    increment = 0
    if salary < 5:
        if rating == 'A':
            increment = 16
        elif rating == 'B':
            increment = 12
        elif rating == 'C':
            increment = 10
        elif rating == 'D':
            increment = 6
    elif salary < 10:
        if rating == 'A':
            increment = 14
        elif rating == 'B':
            increment = 10
        elif rating == 'C':
            increment = 8
        elif rating == 'D':
            increment = 6
    elif salary < 15:
        if rating == 'A':
            increment = 8
        elif rating == 'B':
            increment = 6
        elif rating == 'C':
            increment = 4
        elif rating == 'D':
            increment = 0
    elif salary < 23:
        if rating == 'A':
            increment = 7
        elif rating == 'B':
            increment = 5
        elif rating == 'C':
            increment = 4
        elif rating == 'D':
            increment = 0
    else:
        print("Salary is above defined range")
        exit()

    new_salary = salary + (salary * increment / 100)

    print("Increment percentage:", increment, "%")
    print("New salary:", new_salary)

else:
    print("Please enter valid salary")
```

13.	Ask user to opt for courses for master degree based on the following
L1 = [“HR”, “Finance”, “Marketing”, “DS”]
Based on above subject there are two different streams. For example- HR is having HR core and HR analytics and Marketing is having core and Marketing analytics. Analytics is the optional subject and having added extra fees. DS is not having analytics.
If fees for L1 is 2 lakhs for each course core subject having the same fees but analytics subject having 10% extra on 2 lakhs.
If student opts for hostel then 2 lakhs per year is added. For food monthly 2000 .
Transportation charges 13000 per semester. Calculate the total annual cost based on selected service.  
User will enter values as subject, analytics(Y/N), Hostel (Y/N), food(How many months?), Transportation(semester/annual)


```python
#Method-1-if else
core_fee=200000
analytics_extra_percent=0.10
Hostel_fee=200000
food_per_month=2000
transportation_per_semester=13000
l1=['HR','Finance','Marketing','DS']
subject=input("Which course you want to pursue(HR,Finance,Marketing,DS)").upper()
if subject not in  l1:
  print("Course is not available:")
else:
  analytics=input("Do you want analytics?(Y/N) :").upper()
  hostel=input("Do you want hostel?(Y/N) :").upper()
  food_months=int(input("Enter no. of foods months:"))
  transport=input("Do you want transportation facility?(semester/Annual) :").upper()
  course_fee=core_fee
  if subject !="DS" and analytics=='Y':
    analytics_fee=core_fee*analytics_extra_percent
    course_fee+=analytics_fee
  else:
    analytics_fee=0
  if hostel=='Y':
    Hostel_fee=Hostel_fee 
  else:
    Hostel_fee=0
  food_fee=food_per_month*food_months
  if transport=='SEMESTER':
    transport_fee=transportation_per_semester
  elif transport=='ANNUAl':
    transport_fee=transportation_per_semester*2
  else:
    print("Invalid transportation option")
    transport_fee=0
  total_cost=course_fee+Hostel_fee+food_fee+transport_fee 
  print("course_fee: ",course_fee)
  print("Hostel_Fee: ",Hostel_fee)
  print("food_fee: ",food_fee)
  print("Transport_Fee: ",transport_fee)
  print("-"*50)
  print("Total Annual Cost:",total_cost)
```

```python
#Using OOPs
class DegreeFee:
    def __init__(self):
        self.core_fee=200000
        self.analytics_extra_percent=0.10
        self.hostel_fee_per_year=200000
        self.food_per_month=2000
        self.transportation_per_semester=13000
        self.l1=['HR', 'FINANCE', 'MARKETING', 'DS']
        self.subject=""
        self.analytics=""
        self.hostel=""
        self.food_months=0
        self.transport=""
        self.analytics_fee=0
        self.course_fee=0
        self.hostel_fee=0
        self.food_fee=0
        self.transport_fee=0
        self.total_cost=0
    def take_input(self):
        self.subject=input("Which course you want to pursue (HR, Finance, Marketing, DS): ").upper()
        if self.subject not in self.l1:
            print("Course is not available")
            exit()
        self.analytics=input("Do you want analytics? (Y/N): ").upper()
        self.hostel=input("Do you want hostel? (Y/N): ").upper()
        self.food_months=int(input("Enter number of food months: "))
        self.transport=input("Transportation (SEMESTER/ANNUAL): ").upper()
    def calculate_fees(self):
        self.course_fee=self.core_fee
        if self.subject!= "DS" and self.analytics =='Y':
            self.analytics_fee =self.core_fee*self.analytics_extra_percent
            self.course_fee += self.analytics_fee
        else:
            self.analytics_fee=0
        if self.hostel=='Y':
            self.hostel_fee=self.hostel_fee_per_year
        else:
            self.hostel_fee=0
        self.food_fee=self.food_per_month*self.food_months
        if self.transport=='SEMESTER':
            self.transport_fee=self.transportation_per_semester
        elif self.transport=='ANNUAL':
            self.transport_fee=self.transportation_per_semester*2
        else:
            print("Invalid transport option")
            exit()
        self.total_cost=(self.course_fee+self.hostel_fee+self.food_fee+self.transport_fee)
    def print_bill(self):
        print("Course Selected     :",self.subject) 
        print("Analytics Fee (10%) :",self.analytics_fee)
        print("Course Fee          :",self.course_fee) 
        print("Hostel Fee          :",self.hostel_fee) 
        print("Food Fee            :",self.food_fee) 
        print("Transportation Fee  :",self.transport_fee)
        print("-"*50) 
        print("Total Annual Cost   :",self.total_cost)
obj = DegreeFee()
obj.take_input()
obj.calculate_fees()
obj.print_bill()
```

```python
#using Function
Stationary_Bill = {
    "Books": {
        "1st-4th": {"Maths": 200, "English": 180},
        "5th-8th": {"Maths": 250, "Science": 300},
        "9th-10th": {"Physics": 400, "Chemistry": 380}
    },
    "Notebooks": {
        "square": {100: 40, 200: 70},
        "single": {100: 35, 200: 60},
        "A4 Notebook": {200: 90, 300: 120}
    }
}
def take_order():
    bill = {}
    choice = input("What you want (books/notebooks): ").lower()
    if choice == "books":
        bill["Item"] = "Book"
        print("Available standards:", list(Stationary_Bill["Books"].keys()))
        std = int(input("Enter standard: "))
        if std in [1, 2, 3, 4]:
            key = "1st-4th"
        elif std in [5, 6, 7, 8]:
            key = "5th-8th"
        elif std in [9, 10]:
            key = "9th-10th"
        else:
            print("Invalid standard")
            return None
        print("Available subjects:", list(Stationary_Bill["Books"][key].keys()))
        sub = input("Enter subject: ").title()
        if sub not in Stationary_Bill["Books"][key]:
            print("Invalid subject")
            return None
        price = Stationary_Bill["Books"][key][sub]
        quantity = int(input("Enter quantity: "))
        total = price * quantity
        bill.update({
            "Standard": std,
            "Standard_group": key,
            "Subject": sub,
            "Price_per_unit": price,
            "Quantity": quantity,
            "Total_amount": total
        })
    elif choice == "notebooks":
        bill["Item"] = "Notebook"
        print("Available notebooks:", list(Stationary_Bill["Notebooks"].keys()))
        notebook_type = input("Enter notebook type: ")
        if notebook_type not in Stationary_Bill["Notebooks"]:
            print("Invalid notebook type")
            return None
        print("Available pages:", list(Stationary_Bill["Notebooks"][notebook_type].keys()))
        pages = int(input("Enter pages: "))
        if pages not in Stationary_Bill["Notebooks"][notebook_type]:
            print("Invalid page option")
            return None
        price = Stationary_Bill["Notebooks"][notebook_type][pages]
        quantity = int(input("Enter quantity: "))
        total = price * quantity
        bill.update({
            "Notebook_Type": notebook_type,
            "Pages": pages,
            "Price_per_unit": price,
            "Quantity": quantity,
            "Total_amount": total
        })
    else:
        print("Invalid choice")
        return None
    return bill
def print_bill(bill):
    print("\n" + "=" * 40)
    for key, value in bill.items():
        print(f"{key} : {value}")
    print("=" * 40)
place_order = input("Do you want to place order? (yes/no): ").lower()
if place_order == "yes":
    bill_data = take_order()
    if bill_data:
        print_option = input("Do you want to print the bill? (yes/no): ").lower()
        if print_option == "yes":
            print_bill(bill_data)
        else:
            print("Order placed successfully.")
else:
    print("Thank you.")
```

```python
#46.Create a function to calculate age till now.
import datetime
from datetime import date,datetime
def age_calculator(dob):
    try:
        DOB=datetime.strptime(dob,"%d-%m-%Y").date()
        today=date.today()
        age=today.year-DOB.year
        if(today.year,today.month)<(DOB.year,DOB.month):
            age -=1
        return age
    except ValueError:
        print("Invalid date Format")
```

```python
print(age_calculator("30-10-2003"))
```

```python
#47.Create a function to check age eligibility for given customer based on DOB. Function will take two input DOB and ELIGIBILITY age.
def check_age_eligibility(dob,eligibility_age):
    if isinstance(dob,str):
        dob=datetime.strptime(dob,"%d-%m-%Y").date()
    today=date.today()
    age=today.year-dob.year
    if (today.month,today.day)<(dob.month,dob.day):
        age -=1
        if age >=eligibility_age:
            return f"Eligible(Age:{age})"
        else:
            return f"Not Eligible(Age: {age})"
        
```

```python
print(check_age_eligibility("30-10-2003",25))
```

```python
#49.Create a function to check if string is palindrome or not ? For example, if input is NITIN then reverse of the string is same then it is palindrome. If input is ANIL then reverse is LINA which is not same then it is not palindrome.  
def palindrome(str1):
    if str1==str1[::-1]:
        print(f"{str1} is palindrome")
    else:
        print(f"{str1} is not palindrome")
```
